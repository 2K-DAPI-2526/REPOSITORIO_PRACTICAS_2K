Ejercicio 1

Escribir una función que pida un número entero entre 1 y 10 y guarde en un fichero con el nombre tabla-n.txt la tabla de multiplicar de ese número, donde n es el número introducido.

```python
#Escribir una función que pida un número entero 1 y 10 y guarde en un fichero con el nombre tabla-n.txt (siendo n el número introducido) la tabla de multiplicar de ese número hasta el 10. Por ejemplo, si se introduce un 5 se creará un fichero tabla-5.txt con el siguiente contenido:
#5 x 1 = 5
#5 x 2 = 10
#...    
def guardar_tabla_multiplicar():
    numero = int(input("Introduce un número entero entre 1 y 10: "))
    if 1 <= numero <= 10:
        nombre_fichero = f"tabla-{numero}.txt"
        with open(nombre_fichero, 'w') as fichero:
            for i in range(1, 11):
                resultado = numero * i
                fichero.write(f"{numero} x {i} = {resultado}\n")
        print(f"Tabla de multiplicar del {numero} guardada en {nombre_fichero}.")
    else:
        print("Número fuera de rango. Por favor, introduce un número entre 1 y 10.")
guardar_tabla_multiplicar()

    
```

```python
#Escribir una función que pida un número entero 1 y 10 y guarde en un fichero con el nombre tabla-n.txt (siendo n el número introducido) la tabla de multiplicar de ese número hasta el 10. Por ejemplo, si se introduce un 5 se creará un fichero tabla-5.txt con el siguiente contenido. Si el fichero ya existe, se debe añadir la tabla al final del fichero sin borrar el contenido anterior.

def guardar_tabla_multiplicar_append():
    numero = int(input("Introduce un número entero entre 1 y 10: "))
    if 1 <= numero <= 10:
        nombre_fichero = f"tabla-{numero}.txt"
        with open(nombre_fichero, 'a') as fichero:
            for i in range(1, 11):
                resultado = numero * i
                fichero.write(f"{numero} x {i} = {resultado}\n")
        print(f"Tabla de multiplicar del {numero} añadida en {nombre_fichero}.")
    else:
        print("Número fuera de rango. Por favor, introduce un número entre 1 y 10.")
        
```
