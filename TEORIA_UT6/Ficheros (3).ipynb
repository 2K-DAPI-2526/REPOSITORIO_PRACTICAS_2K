Escribir una función que pida un número entero entre 1 y 10 y guarde en un fichero con el nombre tabla-n.txt la tabla de multiplicar de ese número, donde n es el número introducido.

```python
#Escribir una función que pida un número entero 1 y 10 y guarde en un fichero con el nombre tabla-n.txt (siendo n el número introducido) la tabla de multiplicar de ese número hasta el 10. Por ejemplo, si se introduce un 5 se creará un fichero tabla-5.txt con el siguiente contenido:
#5 x 1 = 5
#5 x 2 = 10
#...    
def guardar_tabla_multiplicar():
    numero = int(input("Introduce un número entero entre 1 y 10: "))
    if 1 <= numero <= 10:
        nombre_fichero = f"tabla-{numero}.txt"
        with open(nombre_fichero, 'w') as fichero:
            for i in range(1, 11):
                resultado = numero * i
                fichero.write(f"{numero} x {i} = {resultado}\n")
        print(f"Tabla de multiplicar del {numero} guardada en {nombre_fichero}.")
    else:
        print("Número fuera de rango. Por favor, introduce un número entre 1 y 10.")
guardar_tabla_multiplicar()

    
```

```python
import pandas as pd
#Para que sirve importar el pandas as pd?
#Sirve para trabajar con estructuras de datos como DataFrames y Series, facilitando el análisis y manipulación de datos tabulares.

# --- 1. Función para leer el CSV y devolver un diccionario por columnas ---
def leer_cotizaciones_a_diccionario(ruta_fichero): 
    #Lee un fichero CSV de cotizaciones y devuelve un diccionario con los datos por columnas
    #Args:
    #    ruta_fichero (str): La ruta al archivo cotizacion.csv.
    #Returns:
    #    dict: Un diccionario donde las claves son los nombres de las columnas
    #          y los valores son listas de los datos de esa columna.
    try:
        # Lee el CSV directamente a un DataFrame de pandas
        df = pd.read_csv(ruta_fichero)
        # Convierte el DataFrame a un diccionario de columnas
        datos_por_columnas = df.to_dict(orient='list')
        return datos_por_columnas
    except FileNotFoundError:
        print(f"Error: El archivo '{ruta_fichero}' no fue encontrado.")
        return None
    except Exception as e:
        print(f"Ocurrió un error al leer el archivo: {e}")
        return None

# --- 2. Función para procesar el diccionario y crear un nuevo CSV ---
def procesar_y_guardar_estadisticas(datos_diccionario, nombre_archivo_salida="estadisticas_cotizaciones.csv"):
    #Recibe el diccionario de datos y crea un CSV con el mínimo, máximo y media de las columnas numéricas.
    #Args:
    #    datos_diccionario (dict): Diccionario de datos por columnas.
    #   nombre_archivo_salida (str): Nombre del archivo CSV de salida.
    if not datos_diccionario:
        print("No hay datos para procesar.")
        return

    # Convierte el diccionario de vuelta a un DataFrame para facilitar el cálculo
    df = pd.DataFrame(datos_diccionario)

    # Define las columnas numéricas a procesar (excluyendo 'Nombre')
    columnas_numericas = ['Final', 'Máximo', 'Mínimo', 'Volumen', 'Efectivo']

    # Diccionario para almacenar los resultados
    resultados = {
        'Metrica': [],
        'Minimo': [],
        'Maximo': [],
        'Media': []
    }

    for col in columnas_numericas:
        if col in df.columns:
            # Asegurarse de que la columna es numérica, si no, convertir (manejo básico)
            df[col] = pd.to_numeric(df[col], errors='coerce') # Convierte a numérico, los no válidos a NaN
            
            # Calcular estadísticas solo si hay datos numéricos válidos
            if not df[col].isnull().all():
                resultados['Metrica'].append(col)
                resultados['Minimo'].append(df[col].min())
                resultados['Maximo'].append(df[col].max())
                resultados['Media'].append(df[col].mean())
            else:
                print(f"Advertencia: La columna '{col}' no contiene datos numéricos válidos.")

    # Crear DataFrame de resultados
    df_resultados = pd.DataFrame(resultados)

    # Guardar en un nuevo archivo CSV
    df_resultados.to_csv(nombre_archivo_salida, index=False)
    print(f"Estadísticas guardadas en '{nombre_archivo_salida}'")

# --- Ejemplo de uso ---
# 1. Crea un archivo 'cotizacion.csv' de prueba (necesitas tener uno real)
#    Para probar, puedes crear uno manualmente o simularlo así:
datos_ejemplo = {
    'Nombre': ['EmpA', 'EmpB', 'EmpC', 'EmpD'],
    'Final': [10.5, 20.2, 5.8, 15.0],
    'Máximo': [11.0, 21.0, 6.0, 15.5
```
