**HERRAMIENTA PYGAME - GUÍA E INSTALACIÓN**

**¿Qué es Pygame?**

Pygame no es un juego, sino un framework (herramienta) para crear juegos en Python.

Permite manejar gráficos, sonido, entradas de teclado y ratón de forma sencilla.

No trae juegos completos listos, pero incluye ejemplos para aprender.

**Cómo instalar Pygame**

Abre tu terminal o consola.

Verifica tu versión de Python:

```python
python --version
```

Instala Pygame con:

```python
python -m pip install pygame
```

o bien,

```python
py -m pip install pygame
```

Verifica que funciona con un ejemplo mínimo:

```python
import pygame
pygame.init()
print("¡Pygame funciona correctamente!")
```

Pygame trae ejemplos de juegos y demos para practicar:


```python
pygame.examples.aliens  #Juego tipo Space Invaders (clásico)
pygame.examples.stars   #Campo de estrellas moviéndose
pygame.examples.chimp   #Monkey que se mueve y puedes golpearlo
pygame.examples.moveit  #Objetos animados moviéndose
pygame.examples.fonty   #Muestra texto con varias fuentes
pygame.examples.vgrade  #Muestra un degradado vertical
pygame.examples.eventlist   #Lista eventos de teclado/ratón
pygame.examples.arraydemo   #Demuestra surfarray
pygame.examples.sound   #Carga y reproduce sonido
pygame.examples.sound_array_demos   #Demos avanzados de sonido
```
